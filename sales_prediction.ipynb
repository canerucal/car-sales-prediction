# Veri Ön-İşleme

```python
#gereksiz olan sütunların veriden çıkarılması işlemi
df = df.drop(['Model','Latest_Launch'], axis = 1)
df.describe(include="all")
```

```python
#boş veri içeren sütunların tespiti
df.isnull().sum()
```

```python
#boş değerlerin veriden silinmesi
df = df.dropna(axis=0)
df.describe(include="all")
```

```python
#silinenlerden sonra verinin büyüklüğü
df.shape
```

```python
#binlik olarak satış rakamları
sns.displot(df['Sales_in_thousands'])
```

```python
#veride aykırıların (outlier) silinmesi
q = df['Sales_in_thousands'].quantile(0.99)
data_1 = df[df['Sales_in_thousands']<q]
data_1.describe(include="all")
```

```python
#beygir gücü sütunundan aykırıların (outlier) silinmesi
sns.distplot(df['Horsepower'])
q = df['Horsepower'].quantile(0.99)
data_2 = df[df['Horsepower']<q]
data_2.describe(include="all")
```

```python
#temizlenmiş veri
data_cleaned = data_2.reset_index(drop=True)
data_cleaned.describe(include="all")
```

```python
#yeni verinin başlıkları
data_cleaned.columns.values
```

```python
#varyans enflasyon faktörünün(VIF) hesaplanması
from statsmodels.stats.outliers_influence import variance_inflation_factor
x = data_cleaned[['Horsepower','Fuel_capacity',
       'Fuel_efficiency', 'Power_perf_factor']]

#VIF verisi
vif_data = pd.DataFrame()
vif_data['feature'] = x.columns

#her sütun için VIF hesaplanması
vif_data['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
print(vif_data)
```

```python
#diğer kalan sütunlar için de VIF hesaplanması
x = data_cleaned[['Wheelbase', 'Width', 'Length', 'Curb_weight']]

#VIF verisi
vif_data = pd.DataFrame()
vif_data['feature'] = x.columns

#her sütun için VIF hesaplanması
vif_data['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
print(vif_data)
```

```python
#lüzumsuz sütunların veriden atılması
data_with_no_collinear = data_cleaned.drop(['Horsepower','Fuel_capacity','Power_perf_factor','Wheelbase', 'Width', 'Length', 'Curb_weight'],axis=1)
data_with_no_collinear.head()
```

```python
#dummy değişkenin oluşturulması
data_with_dummies = pd.get_dummies(data_with_no_collinear, drop_first=True)
data_with_dummies.head()
```

```python
#dummy değişkenli veri
data_with_dummies.columns.values
```

```python
#veriye ince ayar yapılması
cols = ['Price_in_thousands','Sales_in_thousands', '__year_resale_value',
       'Engine_size', 'Fuel_efficiency', 'Manufacturer_Audi',
       'Manufacturer_BMW', 'Manufacturer_Buick', 'Manufacturer_Cadillac',
       'Manufacturer_Chevrolet', 'Manufacturer_Chrysler',
       'Manufacturer_Dodge', 'Manufacturer_Ford', 'Manufacturer_Honda',
       'Manufacturer_Hyundai', 'Manufacturer_Infiniti',
       'Manufacturer_Jeep', 'Manufacturer_Lexus', 'Manufacturer_Lincoln',
       'Manufacturer_Mercedes-B', 'Manufacturer_Mercury',
       'Manufacturer_Mitsubishi', 'Manufacturer_Nissan',
       'Manufacturer_Oldsmobile', 'Manufacturer_Plymouth',
       'Manufacturer_Pontiac', 'Manufacturer_Porsche',
       'Manufacturer_Saturn', 'Manufacturer_Toyota',
       'Manufacturer_Volkswagen', 'Vehicle_type_Passenger']
```

```python
#ön-işlenmiş veri
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

# Makine Öğrenmesi Modeli

```python
#girdi ve hedef verisetlerinin oluşturulması
targets = data_preprocessed['Price_in_thousands']
inputs = data_preprocessed.drop(['Price_in_thousands'],axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
#verinin ölçeklendirilmesi (scale)
sc = StandardScaler()
sc.fit(inputs)
```

```python
input_scaled = sc.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split

#eğitim ve test verisinin bölünmesi
x_train,x_test,y_train,y_test = train_test_split(input_scaled,targets,test_size = 0.2, random_state=42)
```

```python
#regresyon oluşturulması
#eğitim seti için modelin fit edilmesi
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#eğitim veriseti için değerlerin tahmin edilmesi
y_hat = model.predict(x_train)
```

```python
#Hedef ve tahminlerin bastırılması
plt.scatter(y_train,y_hat)
plt.xlabel("Hedefler (y_train)", size = 18)
plt.ylabel("Tahminler (y_hat)", size = 18)
plt.show()
```

```python
sns.displot(y_train - y_hat)
plt.title("Kalıntı Olasılık Dağılım Fonksiyonu (Residual PDF)", size = 18)
```

```python
#model skorunun bastırılması (eğitim için)
model.score(x_train, y_train)
```

```python
#veritabanınındaki sütunlarının ağırlıklarının bastırılması
model_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
model_summary['Weights'] = model.coef_
model_summary
```

```python
#modelin test edilmesi
y_hat_test = model.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test)
plt.xlabel("Hedefler (y_test)", size = 18)
plt.ylabel("Tahminler (y_hat_test)", size = 18)
plt.show()
```

```python
#tahmin edilen satış fiyatının bastırılması
df_pf = pd.DataFrame(y_hat_test,columns=['Tahmin Edilen Satış Fiyatı'])
df_pf.head()
```

```python
#tahmin edilen satış fiyatı ve gerçek satış fiyatının karşılaştırılması
y_test = y_test.reset_index(drop=True)
df_pf['Gerçek Satış Fiyatı'] = y_test
df_pf
```

```python
#tahmin ve gerçeğin grafiğe dökülmesi
sns.lmplot('Tahmin Edilen Satış Fiyatı','Gerçek Satış Fiyatı',data=df_pf)
```
